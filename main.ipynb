```python
from src.IrisData import IrisData
data = IrisData()
```


# k-means clustering

```python
from src.KMeansClustering import *
kMeans = KMeansClustering(data, k = 3, maxItr = 100)
kMeans.run()
```

```python
import matplotlib.pyplot as plt
plt.plot(kMeans.objFunc, "o-r")
plt.ylabel("Objective Function")
plt.xlabel("Iteration")
plt.show()
```

```python
graphShaded = False
if graphShaded:
    for i in range(2, 6):
        print(f"k = {i}")
        kMeans = KMeansClustering(data, k = i, maxItr = 100)
        kMeans.run(withFigure=False)
        _ = kMeans.plotClusters()
        _ = kMeans.plotClusters(shading=True)
```

```python
from src.NeuralNetwork import *
```

```python
dataset = data.getData()
dataset.switchToNeuralNetwork()

network = NeuralNetwork(weights=np.asarray([-0.7, -0.5,  1.9,  1.5]))
Training.plotClassification(dataset.getPoints(), network)
records = Training.train(dataset.getPoints(), network, dataset.getLabels(), 1000)
Training.plotClassification(dataset.getPoints(), network)

network = NeuralNetwork(weights=np.asarray([-3.978, -3.467,  5.316, 5.326]))
Training.plotClassification(dataset.getPoints(), network)

iterations, errors = zip(*records)
fig, ax = plt.subplots()
ax.plot(iterations, errors)
ax.set_title("MSE vs Iteration")
ax.set_xlabel("Iteration")
ax.set_ylabel("MSE")
```
